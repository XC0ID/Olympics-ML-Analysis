### Imports sys

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt

from src.data.loader    import load_all
from src.data.cleaner   import clean_medals, clean_countries
from src.data.validator import validate_medals, validate_countries
```

### Load raw data

```python
medals_raw, countries_raw = load_all(config_path='../config/config.yaml')
print('Raw medals shape   :', medals_raw.shape)
print('Raw countries shape:', countries_raw.shape)
```

### Validate raw data

```python
validate_medals(medals_raw)
validate_countries(countries_raw)
```

### Check nulls before cleaning

```python
print('--- Nulls BEFORE cleaning ---')
print(medals_raw.isnull().sum())
```

### Check duplicates before cleaning

```python
print('Duplicate rows in medals:', medals_raw.duplicated().sum())
```

### Check invalid medal values

```python
print('Unique medal values:', medals_raw['Medal'].unique())
```

### Clean medals

```python
medals = clean_medals(medals_raw)
print('Clean medals shape:', medals.shape)
```

### Check nulls after cleaning

```python
print('--- Nulls AFTER cleaning ---')
print(medals.isnull().sum())
```

### Check duplicates after cleaning


```python
print('Duplicate rows after cleaning:', medals.duplicated().sum())
```

### Clean countries


```python
countries = clean_countries(countries_raw)
print('Clean countries shape:', countries.shape)
countries.head()
```

### Year range check


```python
print('Year range:', medals['Year'].min(), '—', medals['Year'].max())
```

### Rows removed during cleaning


```python
removed = len(medals_raw) - len(medals)
print(f'Rows removed during cleaning: {removed}')
print(f'Kept: {len(medals)} / {len(medals_raw)}')
```

### Save cleaned data

```python
medals.to_csv('../data/processed/medals_clean.csv', index=False)
countries.to_csv('../data/processed/countries_clean.csv', index=False)
print('Saved cleaned files to data/processed/')
```
